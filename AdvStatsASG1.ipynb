```python
#Question 3:

#[a]
''' The Probability-Probability Plot (P-P Plot) is essentially a graphical tool in which we plot the theoretical cumulative probabilities v/s the
empirical cumulative probabilities. The P-P Plot is used to assess how well a data set follows a specified distribution, such as the normal distribution.
 
 Here both axes are probabilities belonging to [0,1].'''

''' The Quantile-Quantile Plot (Q-Q Plot) is another graphical tool that helps us compare quantiles of sample distribution against the quantiles of
theoretical reference distribution. The Q-Q Plot is used to determine if a data set follows a particular distribution, such as the normal distribution,
by checking if the points in the plot approximate a straight line.
 
 Here x-axis holds the theoretical quantiles and y-axis holds the sample quantiles.'''
```

```python
#[b]

#We are using np.random.seed(42) to ensure that we get same random numbers everytime we run the code
np.random.seed(42)
# We created 3 sample sizes as mentioned in the question: 10, 100 and 1000
sizes = [10, 100, 1000]
slope, intercept, r = [], [], []
for n in sizes:
    #We've used np.random.normal to generate random numbers from a normal distribution with mean 100  and std deviation 16 with the size of n (10, 100 or 1000)
    data = np.random.normal(loc=100, scale=16, size=n)
    
    #Ploting the Q-Q plot using scipy's probplot function
    plt.figure()
    stats.probplot(data, dist="norm", plot=plt)
    slope.append(stats.probplot(data, dist="norm")[1][0])
    plt.title(f"Q-Q Plot (n = {n})")
    plt.show()
```

```python
#[c] Slopes:
print("Slopes for n=10, n=100 and n=1000 respectively: ", slope)

''' As sample size keep increasing, we keep getting more clarity with the plot line as well as the slope since more data points helps us find a line
that fits the sample better. Towards n = 1000, we notice that slope is almost equal to std deviation ~ 16. '''
```

```python
#[d]
np.random.seed(42)

#(i) Exponential distribution with scale = 1/2.5
exp = np.random.exponential(scale = 1/2.5, size = 1000)

plt.figure()
stats.probplot(exp, dist="norm", plot=plt)
plt.title("Normal Q-Q Plot for Exponential Distribution scale = 1/2.5")
plt.show()

#(ii) Uniform distribution between 0 and 1
uniform = np.random.uniform(low = 0, high = 1, size = 1000)
plt.figure()
stats.probplot(uniform, dist="norm", plot=plt)
plt.title("Normal Q-Q Plot for Uniform Distribution")
plt.show()
```

```python
'''As clearly seen, in the normal quantile plot for exponential distribution, the points do not form a straight line like a normal distribution would,
instead they form a curve, to be specific a concave one. Hence we can say that exponential distribution's normal quantile plot follows non normality. '''

''' Now, talking about the normal quantile plot for uniform distribution, we see that it's an S-Shaped curve, which is also not a straight line.
 Hence we can say that uniform distribution's normal quantile plot also follows non normality.'''

''' In short :- Both do not form a straight line, hence both do not follow normality with their normal quantile plots.'''
```

```python
#[e]

#Here we are constructing a uniform q-q plot for the uniform distribution
plt.figure()
stats.probplot(uniform, dist="uniform", plot=plt)
plt.title("Uniform Q-Q Plot (Uniform vs Uniform)")
plt.show()
```

```python
''' We notice that when we plot uniform distribution's normal quantile plot, it doesn't form a straight line, but when we plot 
uniform distribution's uniform quantile plot, it forms a straight line. This is because in the first case, we are comparing the quantiles 
of uniform distribution with the quantiles of normal distribution, which are different distributions, hence it doesn't form a straight line. 
But in the second case, we are comparing the quantiles of uniform distribution with the quantiles of uniform distribution itself, which are same 
distributions, hence it forms a straight line. '''
```
